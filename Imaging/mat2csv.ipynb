```python
import scipy.io
import numpy as np
import pandas as pd
import os
```

```python
#list all the mat files and generate a list of filenames
files=os.listdir()
list=files[5:]
#get the number of participants
subj_num=len(list)
```

```python
#create an empty dataframe for the final table
table=pd.DataFrame()
```

```python
for i in range(subj_num):
    #read filename and load the mat file
    filename=list[i]
    subj=filename.split('_')[1]
    data=scipy.io.loadmat(filename)
    
    #get the cluster names
    name=['']*100
    tmp=[]
    for i in range(100):
        tmp=data['names2'][0,i][0].split('.')
        name[i]=tmp[1]
    
    #get the fisher transformed z scores
    Z=data['Z']

    #convert array to datafram and attach the cluster name
    df=pd.DataFrame(Z, columns=name)
    df['parc']=name

    #convert the dataframe into long format
    df_long=pd.melt(df,id_vars=['parc'],value_vars=name, value_name='z',var_name='parc_2')

    #create a string column for the final header names: clusterxx_clusterxx
    df_long['parc_pair']=df_long['parc']+'_'+df_long['parc_2']
    
    #create two interger columns for the filtering logic test
    df_long['parc2_num']=df_long['parc_2'].str.split("cluster",n=1, expand=True)[1].astype(int)
    df_long['parc_num']=df_long['parc'].str.split("cluster",n=1, expand=True)[1].astype(int)

    #logic test to remove diagonal "NA"s and repeated z scores
    df_long.loc[df_long['parc2_num']>df_long['parc_num'],'filter']=1
    df_long_clean=df_long[df_long['filter']==1]

    #remove unwanted columns
    z_parc = df_long_clean[["parc_pair", "z"]]

    #transpose the long dataframe to wide version, using the parc_pair as header names
    df_wide=z_parc.set_index('parc_pair').T
    df_wide['ID']=subj

    #updating the table after each loop
    frames=[table,df_wide]
    table=pd.concat(frames)
    
    print(subj_num)
```
